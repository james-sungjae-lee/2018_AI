```python
import os
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

# 관련 라이브러리
# 연산을 위한 numpy 와 시각화를 위한 matplotlib 의 pyplot 패키지입니다.
```

```python
def logit(z):
    return 1 / (1 + np.exp(-z))

# 로지스틱 함수를 활성함수로 사용하기 위해 logit 이름의 함수를 정의합니다.
# 이 때, exponential function 을 계산하기 위해 numpy 의 exp 메소드를 사용합니다.
```

```python
def relu(z):
    return np.maximum(0, z)

# relu 함수를 활성함수로 사용하기 위해 relu 이름의 함수를 정의합니다.
# 0과 입력값 중 최대값을 갖도록 계산하기 위해 numpy 의 maximum 메소드를 사용합니다.
```

```python
def derivative(f, z, eps = 0.000001):
    return (f(z + eps) - f(z - eps)) / (2 * eps)

# 함수 f 와 입력값 z 를 받아 미분값을 반환하는 함수입니다.
# 각 함수의 도함수를 그리기 위해 구현한 내용입니다.
```

```python
z = np.linspace(-5, 5, 200)

# 입력인 z 값을 numpy 의 linspace 메소드를 이용하여
# -5 부터 5 사이의 값을 순차적으로 200개 뽑아내어 z 에 저장합니다.
```

### (1) 화면 출력 확인 및 각 활성함수의 특징을 비교 서술

```python
plt.figure(figsize = (11, 4))
plt.plot(z, np.sign(z), "r-", linewidth = 2, label = "step")
plt.plot(z, logit(z), "g--", linewidth = 2, label = "sigmoid")
plt.plot(z, np.tanh(z), "b-", linewidth = 2, label = "tanh")
plt.plot(z, relu(z), "m-.", linewidth = 2, label = "ReLU")
plt.grid(True)
plt.legend(loc = "center right", fontsize = 14)
plt.title("activation function: g(z)", fontsize = 14)
plt.axis([-5, 5, -1.2, 1.2])

plt.show()

# 위의 코드는 step 함수, logistic sigmoid 함수, hyperbolic tangent 함수, ReLU 함수를 그립니다.

# 먼저 step 함수는 입력값을 -1또는 1로 맵핑하는 함수로, 퍼셉트론 분류에 사용됩니다.
# step 함수는 0지점에서는 미분이 불가하며, 그 외에서는 모두 미분값이 0으로 딱딱한 의사결정을 하게 됩니다.

# 다음으로 logistic sigmoid 함수는 입력값을 0부터 1사이의 값으로 맵핑하는 함수입니다.
# logistic sigmoid 함수는 모든 구간에서 미분 가능하며, step 함수에 비해 부드러운 의사결정을 합니다.

# 다음으로 hyperbolic tangent 함수인 tanh 은 입력값을 -1부터 1사이의 값으로 맵핑하는 함수입니다.
# tanh 함수는 logistic sigmoid 와 유사하지만 범위가 다르며, 마찬가지로 부드러운 의사결정이 가능합니다.

# 마지막으로 ReLU 함수는 입력값을 0부터 +inf사이의 값으로 맵핑하는 함수입니다.
# ReLU 함수는 sigmoid 와 같이 수렴하는 함수의 단점을 보완하여, Gradient Vanishing 문제를 해결합니다
```

```python
plt.subplot(122)
plt.plot(z, derivative(np.sign, z), "r-", linewidth = 2, label = "step ")
plt.plot(0, 0, "ro", markersize = 5)
plt.plot(0, 0, "rx", markersize = 10)
plt.plot(z, derivative(logit, z), "g--", linewidth = 2, label = "sigmoid")
plt.plot(z, derivative(np.tanh, z), "b-", linewidth = 2, label = "tanh")
plt.plot(z, derivative(relu, z), "m-.", linewidth = 2, label = "ReLU")
plt.grid(True)
plt.title("gradient : g'(z)", fontsize = 14)
plt.legend(fontsize = 7)
plt.axis([-5, 5, -0.2, 1.2])
plt.show()

# 위의 4개 함수에 대하여 각각의 도함수를 그래프로 표현한 것입니다. 위에서 정의한 derivative 함수를 사용합니다
# 먼저 step 함수는 0을 제외한 모든 구간에서 도함수의 값이 0을 갖게 됩니다. 
# sigmoid 함수와 tanh 함수는 유사한 형태의 도함수를 갖지만, tanh 함수의 범위가 더 넓습니다
# 마지막으로 ReLU 함수는 도함수가 0을 기준으로 0 또는 1값을 갖게 됩니다.
```
